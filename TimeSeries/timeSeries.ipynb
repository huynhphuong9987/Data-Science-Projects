- Time Series is said to be stationary when the time series data has __const. mean & variance__ over time. Also, the __correlation between the observations (yt,yt-1) is independent of time.__
- If the time series is not stationary, we will have to make it stationary (as most of the models like ARIMA assume time series to be stationary).
- Also, stationary time series is easier to forecast as we know its statistical properties and they will tend to remain const over future.
- There are three ways to check the stationarity of time series, we will explore all three ways here :
    1. __Plotting ACF & PACF plots__ : If the time series is stationary, the ACF/PACF plots will show a __quick drop-off__ after certain lags.
    2. __Rolling Mean & Std. Deviation__ : Moving Avg. and Moving Std. Dev. is const. over time for a stationary time series.
    3. __Augmented Dickey Fuller test__ : It is a statistical test for checking stationarity of time series. 
      - Null Hypothesis(H0) : TS is non-stationary
      - Alternate Hypothesis(H1) : TS is stationary
      - If p-value < critical-value, we will reject null hyp. and accept alt. hyp.



## 3. Testing for stationarity

#### 3.1 Plotting ACF and PACF


```python
# ACF & PACF Plots
plt.figure()
plt.subplot(211)
plot_acf(new_data.total, ax=plt.gca(), lags = 30)
plt.subplot(212)
plot_pacf(new_data.total, ax=plt.gca(), lags = 30)
plt.show()
```

- ACF is the plot of correlation between y(t) and y(t-k) for lag k
- PACF is the plot of correlation between y(t) and y(t-k) with intermediate lags removed.
- The shaded cone in both the plots is the confidence interval. By default, it is 95% CI or 5% significance level.
- ACF plot shows quick drop off or back up in correlation, hence we can infer that the TS is stationary

#### 3.2 Rolling mean and std. deviation

```python
# Rolling Mean & Rolling Standard Deviation
rolmean = new_data.total.rolling(window = 12).mean()
rolstd = new_data.total.rolling(window = 12).std()

plt.figure(figsize = (15,6))
orig = plt.plot(new_data.total, color = 'blue', label ='Original')
mean  = plt.plot(rolmean, color = 'red', label = 'Rolling Mean')
std = plt.plot(rolstd, color = 'black', label = 'Rolling Std. Dev.')
plt.legend(loc = 'best')
plt.show()
```

Although there is no explicid trend in the data, there may exist the seasonality in the data. In fact, the moving average jumps up and down periodically. It might the pattern thoughout the days within a week.

#### 3.3 ADF test

```python
# test_result=adfuller(new_data['total'])
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(total):
    result=adfuller(total)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("Weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")

adfuller_test(new_data['total'])
```

We can see that the test rejects the null hypothesis and shows the strong evidence that there is not stationarity in the data. 
Since the data is stationary, we can start building an ARIMA model.

## 6. Build ARIMA model

#### 6.1 Splitting the data

```python
train_data = new_data[:350]
test_data = new_data[350:]
train_data
```

#### 6.2 Estimating p, q, and d parameters from ACF and PACF plot.

#### ARIMA(p,d,q)
- p is AR, the number of autoregressive terms. Autoregression is a time series model that uses observations from previous time steps as input to a regresion equation to predict the vlaue at the next time step.  For p value, look at PACF plot.
- q is MA terms, the number of lagged forecast errors in the prediction equation (the size of the moving average window, also known as the order of the MA). For q value, look at ACF plot
- For d, we can check for the stationarity. 
    - d is the number of nonseasonal differences
    - If the data shows a seasonality, add one more d
    - Otherwise, d = 0.

```python
# ACF & PACF Plots
plt.figure()
plt.subplot(211)
# plot_acf(ts_seasonal_diff, ax=plt.gca(), lags = 30)
plot_acf(train_data.total, ax=plt.gca(), lags = 60)
plt.subplot(212)
# plot_pacf(ts_seasonal_diff, ax=plt.gca(), lags = 30)
plot_pacf(train_data.total, ax=plt.gca(), lags = 60)
plt.show()
```

- Plotting the ACF and PACF on the original data:
    - Looking at the ACF, we can see that q = 12 because the 12th lag is the first one that gets very close to the dense part.
    - The same for PACF graph, the p value is 8. 
- d value is 0 since the data is in stationality with no trend or seasonality.

#### 6.3 Model 1 : ARIMA(8,0,12)

```python
model_fit = sm.tsa.arima.model.ARIMA(train_data.total, order = (8,0,12)).fit()
model_fit.summary()
```

```python
predictions1 = model_fit.forecast(steps = len(test_data))
plt.figure(figsize = (20,6))
plt.plot(new_data.total, color = 'green', label = 'Original data')
plt.plot(model_fit.fittedvalues, color = 'blue', label = 'Predicted values for train dataset')
plt.plot(predictions1, color = 'orange', label = 'Predicted values for test dataset')
plt.legend(loc = 'best')
plt.show()
```

```python
MSE_train_1 = mean_squared_error(train_data.total, model_fit.fittedvalues)
print('Squared root of MSE Training model ',np.sqrt(MSE_train_1))
MSE_test_1 = mean_squared_error(test_data.total, predictions1)
print('Squared root of MSE Testing model',np.sqrt(MSE_test_1))
```

#### 6.3 Model 2 : SARIMA => ARIMA(8,0,12)(0,1,0)[52] => seasonal differencing

- ARIMA(p,d,q)(P,D,Q)[m] : m is the number of observations per year. Uppercase notations are for seasonal parts and lowercase are for the non-seasonal part. 

- The seasonal part is similar to the non-seasonal part but involve backshifts of the seasonal period.

- ARIMA(p,d,q)(0,0,1)[m]: exponential decay in the seasonal lags of the PACF (i.e, at lags m, 2m, 3m)

- ARIMA(p,d,q)(1,0,0)[m]: a single significant spike at lag m in the PACF.

- Here, it will be ARIMA(p,d,q)(0,1,0)[52]

```python
model2 = sm.tsa.arima.model.ARIMA(train_data.total,  order = (8,0,12), seasonal_order=(0,1,0,52))
model_fit_2 = model2.fit()
model_fit_2.summary()
```

```python
predictions2 = model_fit_2.forecast(steps = len(test_data))
plt.figure(figsize = (20,6))
plt.plot(new_data.total, color = 'green', label = 'Original data')
plt.plot(model_fit_2.fittedvalues, color = 'blue', label = 'Predicted values for train dataset')
plt.plot(predictions2, color = 'orange', label = 'Predicted values for test dataset')
plt.legend(loc = 'best')
# plt.show()
```

```python
MSE_train_2= mean_squared_error(train_data.total, model_fit_2.fittedvalues)
print('MSE on training data',np.sqrt(MSE_train_2))
MSE_test_2 = mean_squared_error(test_data.total, predictions2)
print('MSE on testing data',np.sqrt(MSE_test_2))
```

## 8. Evaluation

We validate the Time Series model by analyzing the residuals :
1. Residuals should be white noise, i.e, they should be stationary in nature (const. mean & variance over time). We will use ACF & PACF plots for residuals to test this. Quick drop off after certain lags in ACF & PACF plots will confirm it.
2. Residuals should follow normal distribution. We will use histogram and q-q plot to confirm this.
3. There is a statistical test Ljung-Box test which checks whether TS model is a good fit or not using Q-statistic. 
   - H0: Model is a good fit
   - H1: Model is not a good fit
   - Reject H0 if, Q > chi-square(1-alpha,h) (h = m-p-q, p&q are arima model para. and h is max. lag considered) or p-value < alpha

```python
# 1. Checking white noise of residuals
plt.figure(figsize = (15,6))
plt.subplot(3,1,1)
model_fit.resid.plot(kind = 'line',ylabel = 'Residuals', title = 'Residuals plot')
plt.subplot(3,1,2)
plot_acf(model_fit.resid, ax=plt.gca(), lags = 30, title = 'ACF plot of residuals')
plt.subplot(3,1,3)
plot_pacf(model_fit.resid, ax=plt.gca(), lags = 30, title = 'PACF plot of residuals')
plt.show()
```

From ACF and PACF plots, we can observe that almost all lags are 0, it indicates white noise.

```python
# 2. Checking normality of residuals using distribution plot and q-q plot
plt.figure(figsize=(10,5))
sns.histplot(model_fit.resid, kde = True)
plt.title('Distribution plot of residuals')
sm.api.qqplot(model_fit.resid, line = 'r')
plt.title('Q-Q plot of residuals')
plt.show()
```

The residual plots shows a clear normality.

```python
# Ljung-Box residual test on our best model => model => ARIMA(8,0,12)

# For determining max. lags for L-B test, we will use folling formulae :
# h = min(10,n/5) --- for non-seasonal TS
# h = min(2m,n/5) --- for seasonal TS
# n = length of time series
# m = seasonal period

n = len(train_data)
m = 7
# h = min(2*m,n/5)
h = min(10,n/5)
print(h)

sm.stats.diagnostic.acorr_ljungbox(x = model_fit.resid, lags = [h], return_df = True)
```

Since, p-value > 0.05, we accept H0 . Hence, current model is a good fit.
